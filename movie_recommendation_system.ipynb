```python
# Selecting columns that assumes good realtion between features

s_columns=['genres','keywords','tagline','cast','original_language','director']
```

```python
#combining features

c_features=data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['original_language']+' '+data['director']
c_features[0]
```

```python
# converting text data into vectors

from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer()
```

```python
f_vectors =vect.fit_transform(c_features)
f_vectors
```

```python
# Find the similarity score using cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(f_vectors)
similarity
```

```python
similarity.shape
```

```python
# Take a movie in which the user are interested in

movie_name = input('Enter the movie name in which you are interested : ')
```

```python
# creating the list of all the movies

movies_list = data['title'].tolist()
movies_list
```

```python
import difflib

close_matches = difflib.get_close_matches(movie_name, movies_list)
close_matches
```

```python
close_match = close_matches[0]
close_match
```

```python
# Find the index of the movie with title

movie_index = data[data.title == close_match]['index'].values[0]
movie_index
```

```python
similarity_score = list(enumerate(similarity[movie_index]))
similarity_score
```

```python
sort_score=sorted(similarity_score, key = lambda x:x[1], reverse = True)
sort_score
```

```python
# Find the name of the similar movies based on the index

i = 1
for movie in sort_score:
    index = movie[0]
    title_of_the_movie = data[data.index == index]['title'].values[0]
    
    if (i<11):
        print(i, title_of_the_movie)
        i+=1
```
